Access the World Bank database to illustrate Morocco's population and GDP.

世銀データベースにアクセスしてモロッコの人口とGDPを図示する。

```python
#installation of worldbank API https://pypi.org/project/wbgapi/
pip install wbgapi
```

```python
import wbgapi as wb
import pandas as pd
```

```python
#Listing all the series (indicators) of wb
for series in wb.series.list():
    print(f"{series['id']}: {series['value']}")
```

```python
#wb.data.DataFrame('series code','country') how to search the indicator name
s = wb.search('population total')
print(s)
```

```python
#Extract pop data of Morocco as a DF
population = wb.data.DataFrame('SP.POP.TOTL', 'MAR')
print(population)
```

```python
#transpose the rows and columns
dfpopulation = population.transpose()
print(dfpopulation)
```

```python
#visualize the population
dfpopulation['MAR'].plot(x='MAR', kind='line', title='Population (Morocco)')
```

```python
#Display pop and DGP data of Morocco
gdp = wb.data.DataFrame('NY.GDP.PCAP.CD', 'MAR').transpose()
dfgdp = pd.DataFrame(gdp)
dfpopulation['MAR'].plot(x='MAR', kind='line', title='Population vs GDP (Morocco)')
dfgdp['MAR'].plot(x='MAR', kind = 'line', secondary_y=True)
```
